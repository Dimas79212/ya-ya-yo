```python
# Membaca data dari file CSV
df = pd.read_csv('stock.csv')

# Membuat grafik histogram
plt.figure(figsize=(13,6))
plt.hist(df['Purchase Price'], weights=df['Shares'], bins=20, color='aqua', edgecolor='black')

# Membuat x dan y serta judul
plt.title('Histogram dari Purchase Price dan Shares')
plt.xlabel('Purchase Price')
plt.ylabel('Shares')
plt.show()
```

```python
# Membaca data dari file CSV
df = pd.read_csv('stock.csv')

# Pilih data "Dividen / Share" 5 tertinggi
top_dividend_stocks = df.sort_values(by='Dividend/share', ascending=False).head(5)

# Menggambarkan data
plt.figure(figsize=(12,6))
plt.bar(top_dividend_stocks['Stock Name'], top_dividend_stocks['Dividend/share'], color='skyblue', edgecolor='black')
plt.title('5 Tertinggi Dividend/share dengan Stock Name')
plt.xlabel('Stock Name')
plt.ylabel('Dividend/share')
plt.xticks(rotation=45)
plt.show()
```
